##  secao-7:  Dados categoricos

```python
import pandas as pd
import numpy as np

df = pd.read_csv('dados/mtcars.csv')
df.head(2)
```

```python
df.columns = ['nome','mpg','cyl','disp','hp','drat','wt','qsec',
              'vs','am','gear','carb']
df.index = df.nome
```

```python
# Esses sao os dados categoricos dos meu dataframe 
#dados catergoricos sao aqueles que contem apenas algumas categorias
carros = df[['cyl', 'am', 'gear', 'carb']]
carros.head()
```

### vamos fazer um agrupamento com apenas uma variavel categorica para poder comparar com as outras

```python
cyl = carros.groupby('cyl')
cyl.describe().T
```

### Transformando variaveis em dados categoricos

```python
carros['grupo'] = pd.Series(carros['cyl'], dtype='category')
```

```python
carros.info()
```

### quando algo tem o valor category ele nos da uma melhor vizualisaçao das categorisa possiveis 

```python
carros['grupo'].head()
```

#### um metodos mais simples para comparar duas variaveis categoricas é 

```python
pd.crosstab(carros['cyl'], carros['gear'])
```
