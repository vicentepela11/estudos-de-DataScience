```python
import pandas as pd 
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
df = pd.read_csv('dados/mtcars.csv')
df.columns = ['nome','mpg','cyl','disp','hp','drat','wt','qsec',
              'vs','am','gear','carb']
df.index = df.nome
```

```python
# Importando a biblioteca que faz o coeficiente de pears
from scipy.stats import pearsonr
```

## correlaçao de pearson
### ultilizamos essa biblioteca pra fazer correlaçoes entre variaveis e ver se elas realmente se correlacionam umas com as outras porem nao quer dizer que sao causais 
### EX: se exite uma correlaçao entre. uma regiao que tem mercados grande e  uma taxa de obsidade tao grande quanto nao quer dizer que os mercados grandes estao causando obesidade naquela regiao 

```python
x = df[['mpg', 'hp', 'wt', 'qsec']]
# precisamos pegar os dados que tem os 3 requisitos para se fazer uma correlaçao
# 1- apresentar distribuiçao normal 
# 2- variaveis continuas 
# 3- apresentar relaçoes linear 
sns.pairplot(x)
```

# precisamos pegar os dados que tem os 3 requisitos para se fazer uma correlaçao
### 1- apresentar distribuiçao normal (Podemos achar essa distribuiçao com os graficos de histograma)
### 2- variaveis continuas/numericas
### 3- apresentar relaçoes linear  (Podemos achar essa distribuiçao com os graficos de skterplot)

```python
mpg = df['mpg']
hp = df['hp']
qsec = df['qsec']
wt = df['wt']

pearson, p_valor = pearsonr(mpg, hp)
print(f'O coeficiente de pearson {pearson:0.3}')
```

```python
pearson, p_valor = pearsonr(mpg, qsec)
print(f'O coeficiente de pearson {pearson:0.3}')
```

```python
pearson, p_valor = pearsonr(mpg, wt)
print(f'O coeficiente de pearson {pearson:0.3}')
```
