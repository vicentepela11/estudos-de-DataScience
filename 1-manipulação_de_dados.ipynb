## Removendo dados repetidos do DataFrame

```python
#criando dicionario para aplicar no dataframe
dados = {
    "coluna1": [1, 1, 2, 2, 3, 3, 3],
    "coluna2": ['a', 'a', 'b','b', 'c', 'c', 'c'],
    "coluna3": ['A', 'A', 'B','B', 'C', 'C', 'C'],
}
df = pd.DataFrame(dados)
df
```

#### Indentificando os dados duplicados 

```python
df.duplicated()
```

```python
#removendo valores repetido e colocando dentro do meu data frame sem gerar outro
df.drop_duplicates(inplace=True)
df
```

```python
# Mostrando com faz para apenas uma coluna 
dados = {
    "coluna1": [1, 1, 2, 2, 3, 3, 3],
    "coluna2": ['a', 'a', 'b','b', 'c', 'c', 'c'],
    "coluna3": ['D', 'A', 'B','B', 'C', 'C', 'C'],
}
df = pd.DataFrame(dados)
df
```

# A grupamento de dados

```python
df_carro = pd.read_csv("dados/mtcars.csv")
print(df_carro.shape)
df_carro.head()
```

```python
df_carro.columns = ['nome', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb'] 
df_carro
```

```python
df_carro['cyl'].value_counts()
#faremos o agrupamento para esses 3 valores entao veja só 
```

### executando o groupby

```python
#cilindros = df_carro["cyl"]
grupo_cilindros = df_carro.groupby(df_carro["cyl"])
grupo_cilindros.mean()
```
