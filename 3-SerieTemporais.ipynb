## tres tipos de graficos essesenciai para uma boa analise dos dados

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import rcParams

import seaborn as sns

import plotly.express as px

%matplotlib inline
rcParams['figure.figsize'] = 6, 5
sns.set_style("whitegrid")
```

```python
df = pd.read_csv("dados/Superstore-Sales.csv" , index_col='Order Date', parse_dates=True,
                 encoding="ISO-8859-1")
df.head(1)
```

```python
df['Order Quantity'].plot()
```

# como sao muitos dados fica ruim a visualizaçao entao vamos diminuir a quantidade e jogar para outro DataFrame

```python
df2 = df.sample(n=100, random_state=25, axis=0)
df2['Order Quantity'].plot()
plt.xlabel('ordem da data', size=15)
plt.ylabel('ordem de quantidade', size=15)
```

#### Graficos essenciais

```python
dfcarros = pd.read_csv("dados/mtcars.csv")
mpg = dfcarros['mpg']
np.dtype(dfcarros['hp'])
```

```python
#histograma com pyplot
plt.hist(mpg)
plt.show()
```

## vendo os scatteplot

```python
dfcarros.plot(kind="scatter", x='hp', y='mpg', c='red', s=100)
```

# matrizes de scatterplot

```python
# matrizes de scatterplot
sns.pairplot(dfcarros)
```

```python
coluna = ['mpg', 'hp', 'disp', 'wt']
novodf = pd.DataFrame((dfcarros[coluna].values), columns=coluna)
grupo = dfcarros['am'].values
novodf['grupo'] = pd.Series(grupo, dtype='int64')
novodf.head()
```

```python
sns.pairplot(novodf, hue='grupo', palette='hls') 
```

```python
dfcarros.boxplot(column='mpg', by='am')
dfcarros.boxplot(column='wt', by='am')
dfcarros.boxplot(column='hp', by='am')
```

```python
sns.boxplot(x='am', y='wt', data=dfcarros, palette='hls')
```

```python
sns.boxplot(x='am', y='hp', data=dfcarros, palette='hls')
```
